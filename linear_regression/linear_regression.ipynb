# Datetime 

```python
df['datetime'] = pd.to_datetime(df['X1 transaction date'], errors='ignore')
df
```

# Split the data to train and test

```python
home_exp_vals = df['Y house price of unit area'].values
home_exp_vals
```

```python
con_stores = df['X4 number of convenience stores'].values
con_stores
```

```python
home_exp_vector = home_exp_vals.reshape(-1,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(home_exp_vector, con_stores, train_size=0.8, test_size=0.2, random_state=100)
```

```python
print(f"X_train shape {X_train.shape}")
print(f"y_train shape {y_train.shape}")
print(f"X_test shape {X_test.shape}")
print(f"y_test shape {y_test.shape}")
print(y_test)
print(y_train)
```

# Show Score & Model

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
y_predict = lm.predict(X_test)
print(f"Train accuracy {round(lm.score(X_train, y_train)*100,2)} %")
print(f'Test accuracy {round(lm.score(X_test, y_test)*100,2)} %')
```

```python
%matplotlib inline
plt.scatter(X_train, y_train, color="red")
plt.title("trained model plot")
plt.show()
```

```python
num_years = 20
salary_predict = lm.predict([[num_years]])[0]
from sklearn.metrics import r2_score
r2_score(y_test, y_predict)
```
